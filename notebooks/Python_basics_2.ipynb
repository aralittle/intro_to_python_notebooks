### String slicing
You can use string slices to work with substrings within strings

```python
my_string = 'one two three four!'
print(my_string[:]) # print the whole string
print('---')
print(my_string[0:3]) # first index is inclusive, second index is exclusive
print(my_string[:3]) # omiting the first index  starts the slice at the beginning of the string
print('---')
print(my_string[1:]) # omitting the second index extends the slice to the end of the string
print(my_string[-5:-1]) # negative indices can also be used
```

```python
print(my_string[:5] + my_string[5:])
print(my_string[:5] + my_string[5:] == my_string)
```

```python
# Practice: print some substrings from the following string
my_string = 'Colorless green ideas sleep furiously'
# Print 'C' (first character)
print(my_string[:1])

# Print "green"
print(my_string[10:15])

# Use negative indices to print  "furiously"
print(my_string[-9:])

# Print the complete string using slices
print(my_string[:])
```

### Useful methods with strings

#### Replace characters (does not change the variable)

Strings are **inmutable**. This means that they can't be modified. We need to create a new variable to store there the string that we want to modify.


```python
my_string = 'I want to cancel my account\tOKOK'
my_string[-2:] = 'KO'
```

We can replace characters in the string. To do so we can use string slicing or the replace method, which will change all ocurrences of a substring for another substring, or perform n replacements starting from the end if we specify n.

```python
my_string_modified = my_string[:-2] + 'KO'
print(my_string_modified)
my_string_modified2 = my_string.replace('OK', 'KO')
print(my_string_modified2)
```

```python
another_string = "this $ is $ a $ string"
another_string_modified = another_string.replace('$', '', 2)
print(another_string_modified)
```

#### Remove character on the letf/right (does not change the variable)

```python
# remove characters on the left
print('@@hello'.lstrip('@'))
colours_with_spaces = '   blue green yellow   '
colours_with_tabs_and_newlines = '\t\nblue\t\ngreen\t\nyellow\t'
print(colours_with_spaces.lstrip())
print(colours_with_tabs_and_newlines.lstrip())
```

```python
# remove characters on the right
print(colours_with_spaces.rstrip())
print(colours_with_tabs_and_newlines.rstrip())
```

```python
# remove characters on the right and on the left
print(colours_with_spaces.strip())
print(colours_with_tabs_and_newlines.strip())
```

#### Case manipulations (does not change the variable)

```python
sentence = "i'm going to be LATE"
print("capitalization: ", sentence.capitalize())
```

```python
print("to lower case:", sentence.lower())
print("to upper case:",  sentence.upper())
print("swap case:", sentence.swapcase())
print("title case:",  sentence.title()) #it does not handle well apostrophes, possessives, or acronyms
sentence = sentence.title()
print(sentence)
```

#### Character classification

```python
# Determine if a string only contains alphanumeric characters
print('abc123'.isalnum())
print('abc$123'.isalnum())
print(''.isalnum())
```

```python
# Determine if a string only contains alphabetic characters
print('ABCabcñçó'.isalpha())
print('abc123'.isalpha())
print(''.isalpha())
```

```python
# Determine if a string only contains numeric characters
print('123'.isdigit())
print('123abc'.isdigit())
print(''.isdigit())
```

```python
# Determine if the whole string is in lower case
print('abc'.islower())
print('abc1$d'.islower())
print('Abc1$D'.islower())
```

```python
# Determine if the whole string is in upper case
print('ABC'.isupper())
print('ABC1$D'.isupper())
print('Abc1$D'.isupper())
```

#### Strings and lists

Lists are convenient ways to work with strings, because you can work directly with words of other units rather than characters. Besides, they are **mutable**


```python
sentence = "I want to close the account\tCANCEL"
print(sentence)
sentence_list = sentence.split('\t')
print(sentence_list)
```

By default the split method splits strings using whitespace characters (\s\t\n\r...)

```python
sentence_list = sentence.split()
print(sentence_list)
```

You can use indexing to select the elements in the list that you are interested in

```python
sentence_list = sentence.split('\t')
my_sentence = sentence_list[0]
my_intent = sentence_list[1]
print(f"my sentence is '{my_sentence}' and my intent is '{my_intent}'")
```

You can also create a string from a list:

```python
reconstructed_sentence1 = " ".join(sentence_list)
reconstructed_sentence2 = "\t".join(sentence_list)
print(reconstructed_sentence1)
print(reconstructed_sentence2)
```
